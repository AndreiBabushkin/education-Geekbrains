### 1.1.

```python
import pandas as pd
df = pd.read_csv(r'kc_house_data.csv')
df.head()
```

### 1.2
Изучите стоимости недвижимости


```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.hist(df['price'], bins=15)
plt.title('Стоимость недвижимости')
plt.xlabel('Цена недвижимости')
plt.ylabel('Кол-во домов');
```

Вывод: Оснавная масса представленной недвижимости стоит в прелах до 1 млн.

Вывод: Большее количество домов имеют жилую площадь от 200 до 3800 м2

### 1.4

Вывод: В 2000-2010 произошел резкое увеличение застройки.

## 3 задача
Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
import seaborn as sns
plt.figure(figsize=(8, 6))
sns.jointplot(x=df['yr_built'], y=df['price'], kind='reg')
plt.title('Изменение стоимости в зависимости от года постройки')
plt.xlabel('Год постройки')
plt.ylabel('Стоимость');
```

Чем позднее год постройки тем выше стоимоть недвижимости

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x=df['floors'], y=df['price'])
plt.title('Изменение стоимости в зависимости от количества этажей в доме')
plt.xlabel('Этажность')
plt.ylabel('Стоимость');
```

Самые дорогие двух этажные дома, стоимость остальных находится примерно на одном уровне

```python
plt.figure(figsize=(8, 6))
sns.lineplot(x=df['bedrooms'], y=df['price'])
plt.title('Изменение стоимости в зависимости от количества спален')
plt.xlabel('Количество спален')
plt.ylabel('Стоимость');
```

Самыми дорогими являются дома с количеством спален от 5 до 10. Можно отметить линейную зависимость.
